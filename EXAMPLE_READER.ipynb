```python
Ticker = 'WLFC'
CURRENT_MONTH = '2024-10'
```

```python
# Fetch the ticker data
TICKER = yf.Ticker(Ticker)
# Get the EPS estimate
eps_trend = TICKER.eps_trend

# Calculate percent change from 7, 30, 60, 90 days ago to current
eps_trend['7d_change%'] = (eps_trend['current'] - eps_trend['7daysAgo']) / eps_trend['7daysAgo'] * 100
eps_trend['30d_change%'] = (eps_trend['current'] - eps_trend['30daysAgo']) / eps_trend['30daysAgo'] * 100
eps_trend['60d_change%'] = (eps_trend['current'] - eps_trend['60daysAgo']) / eps_trend['60daysAgo'] * 100
eps_trend['90d_change%'] = (eps_trend['current'] - eps_trend['90daysAgo']) / eps_trend['90daysAgo'] * 100

eps_trend
```
